```python
def cost(points,y,m):
    M = len(points)
    total_cost = 0
    for i in range(M):
        x = points[i,:]
        sum=0
        for j in range(points.shape[1]):
            sum += m[j]*x[j]
        total_cost += (1/M)*((y[i]- sum) **2)
            
    return total_cost
```

```python
def stepGrad(points,y,learning_rate,m):
    M = len(points)
    mslope = np.zeros(points.shape[1])
    for i in range(M):
        x = points[i]
        sum=0
        for j in range(len(m)):
            sum += m[j]*x[j]
        for k in range(points.shape[1]):
            mslope[k] += (-2/M)*((y[i]- sum) *x[k])
    mnew = m -  float(learning_rate)*mslope
        
    return mnew
```

```python
def gradDesc(points,y,learning_rate,num_iterations):
    m = np.zeros(points.shape[1])
    for i in range(num_iterations):
        m = stepGrad(points,y,learning_rate,m)
        print(i,"cost:",cost(points,y,m))
    return m
    
```

```python
def run(points,y):
    learning_rate = 0.03
    num_iterations = 1000
    m = gradDesc(points,y,learning_rate,num_iterations)
    return m
```

```python

points = df_train.values[:,:]
y = ytrain
m_ans = run(points,y)
```

```python
ypred = np.zeros(len(bostontest))

for i in range(len(bostontest)):
    x = bostontest[i,:]
    l = len(m_ans)
    tsum = 0
    for j in range(l):
        tsum += m_ans[j]*x[j]
    ypred[i] = tsum
    
np.savetxt("prediction.csv",ypred)
```
